```python
import random
L=[[0]*4 for i in range(4)]
T=[[0]*4 for i in range(4)]
P=[[0]*4 for i in range(4)]
N=[[0]*8 for i in range(2)]


for i in range(4):
  for j in range(4):
    L[i][j]=random.random()


for i in range(4):
  print(L[i],"\n")
print("\n\n\n")


for i in range(4):
  for j in range(4):
    T[j][i]=float(L[i][j])
    


for i in range(4):
  print(T[i],"\n")
print("\n\n\n")

for i in range(4):
  for j in range(4):
      s=0
      for k in range(4):
          s+=(L[i][k]*T[k][j])
      P[i][j]=s
k=0
i=0
while i<2:
  j=0
  while j<4:
    N[0][k]=P[i][j]
    j+=1
    k+=1
  i+=1 
k=0
while i<4:
  j=0
  while j<4:
    N[1][k]=P[i][j]
    j+=1
    k+=1
  i+=1 

for i in range(4):
  print(P[i],"\n")

print("\n\n\n")

for i in range(2):
  print(N[i],"\n")
```
